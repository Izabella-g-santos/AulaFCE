#### **Quest√£o 3**. Fermenta√ß√£o alco√≥lica com cin√©tica dependente da temperatura - Um processo em batelada converte glicose em etanol e CO2. A taxa de consumo da glicose depende da temperatura segundo a express√£o:

A cin√©tica √© de primeira ordem, tal que:

Considerando concentra√ß√£o inicial S(t = 0) de glicose igual a 100 g/L e temperatura constante de 35¬∞C, implemente uma solu√ß√£o computacional para simular a concentra√ß√£o de glicose ao longo de 24 horas. Qual o valor final da concentra√ß√£o de glicose? O que acontece se a temperatura for aumentada para 50¬∞C?


```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from dataclasses import dataclass

@dataclass
class ParametrosBatelada:
    T: float = 35.0     # Temperatura em ¬∞C
    S0: float = 100.0   # Concentra√ß√£o inicial de glicose (g/L)

def modelo_batelada(t, y, params):
    S = y[0]
    T_kelvin = params.T + 273
    k = 464.8 * np.exp(-3985 / T_kelvin)  # [1/min]
    dSdt = -k * S
    return [dSdt]

def simula_batelada(tempo, params):
    teval = np.linspace(tempo[0], tempo[-1], 100)
    sol = solve_ivp(modelo_batelada, tempo, [params.S0], t_eval=teval, method='BDF', args=(params,))
    return sol.t, sol.y[0]

# Simula√ß√µes para T = 35¬∞C e T = 50¬∞C
conjuntos = [
    ParametrosBatelada(T=35.0),
    ParametrosBatelada(T=50.0)
]

tempo = [0, 24*60]  # tempo em minutos

for params in conjuntos:
    t, S = simula_batelada(tempo, params)
    print(f"Temperatura: {params.T} ¬∞C --> Concentra√ß√£o final: S = {S[-1]:.4f} g/L")
    plt.plot(t, S, label=f"T = {params.T} ¬∞C")

plt.xlabel("Tempo (min)")
plt.ylabel("Concentra√ß√£o de glicose (g/L)")
plt.title("Consumo de glicose ao longo de 24h")
plt.legend()
plt.grid()
plt.show()
```

#### **Quest√£o 4. Equil√≠brio em coluna de destila√ß√£o com seletividade dependente da temperatura** - Um processo modelado com tr√™s est√°gios separa uma mistura etanol/√°gua. A fra√ß√£o molar l√≠quida de etanol (xi) e a temperatura (Ti) em cada est√°gio s√£o conhecidas. O equil√≠brio l√≠quido-vapor √© dado por:

Com a seguinte express√£o linear para a seletividade Œ±(T) = 2.5 ‚àí 0.01 ¬∑ (T ‚àí 78).

Os dados para os tr√™s est√°gios s√£o:

Captura de tela 2025-07-23 142018.png

Escreva um c√≥digo modular que calcule e exiba a seletividade Œ±i em cada est√°gio e a fra√ß√£o yi com base nos dados acima.

```python
from dataclasses import dataclass
from tabulate import tabulate

@dataclass
class ParametrosEstagios:
    estagio: str
    xi: float    # Fra√ß√£o molar l√≠quida do etanol
    Ti: float    # Temperatura (¬∞C)

# C√°lculo das fun√ß√µes
def calculo_alpha(Ti):
    return 2.5 - 0.01 * (Ti - 78)

def calculo_yi(xi, Ti):
    alpha = calculo_alpha(Ti)
    return (alpha * xi) / (1 + (alpha - 1) * xi)

# Lista de est√°gios
estagios = [
    ParametrosEstagios("Topo", 0.20, 78),
    ParametrosEstagios("Alimenta√ß√£o", 0.40, 80),
    ParametrosEstagios("Fundo", 0.60, 85)
]

tabela_resultados = []
for estagio in estagios:
    alpha = calculo_alpha(estagio.Ti)
    yi = calculo_yi(estagio.xi, estagio.Ti)
    tabela_resultados.append([
        estagio.estagio,
        f"{estagio.Ti} ¬∞C",
        f"{estagio.xi:.2f}",
        f"{alpha:.4f}",
        f"{yi:.4f}"
    ])

headers = ["Est√°gio", "Temperatura (T·µ¢)", "Fra√ß√£o l√≠quida (x·µ¢)", "Seletividade (Œ±·µ¢)", "Fra√ß√£o no vapor (y·µ¢)"]
print(tabulate(tabela_resultados, headers=headers, tablefmt="grid"))
```
