```python
import numpy as np

i = 0
tol = 1e-6
erro = 100
soma_max = np.pi/4
s = []

while erro > tol:
  soma = (-1)**i/(2*i+1)
  s.append(soma)
  SOMA = np.sum(s)
  erro = (SOMA - soma_max)**2
  i = i + 1
else:
  print(f"\n foram necessárias {i} iterações")
```
